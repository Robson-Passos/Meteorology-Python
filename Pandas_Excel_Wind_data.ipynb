```python
import pandas as pd
import numpy as np
```

```python
excel_file = 'input/SIMA_estreito.xlsx'
dados = pd.read_excel(excel_file, usecols="A,B,C")
dados.shape
```

```python
dados.head()
```

```python
dados = dados.rename(index=str, columns={"Data e horário(UTC)": "data",
                                         "Direção do vento(°NV)": "dir",
                                         "Intensidade do vento(m/s)": "vel"});
dados.head()
```

```python
dados = dados[np.isfinite(dados['vel'])]
dados = dados[np.isfinite(dados['dir'])]
print(dados.shape)
```

```python
from windrose import WindroseAxes, WindAxes, plot_windrose
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 22})  # Isso é para aumentar o valor dos labels N, N-E, E, etc...

ax = WindroseAxes.from_ax()

ax.bar(dados.dir, dados.vel, normed=True, bins=np.arange(0,8,2), opening=0.8, edgecolor='white')

lgd = ax.set_legend(title='Vel. do Vento (m/s)', loc=(1.1, 0), shadow=True)

ax.set_yticks(np.arange(0, 25, step=5))

ax.set_yticklabels(np.arange(0, 25, step=5), fontsize='12')

plt.title("Rosa dos ventos",y=1.08, fontsize='20')

plt.setp(lgd.get_texts(), fontsize=18) # Aumenta o texto da legenda
```

```python
dados = dados[dados['vel'] >= 5].dropna()
dados.head()
```

```python
plt.rcParams.update({'font.size': 22})  # Isso é para aumentar o valor dos labels N, N-E, E, etc...

ax = WindroseAxes.from_ax()

ax.bar(dados.dir, dados.vel, normed=True, bins=np.arange(5,10,1), opening=0.8, edgecolor='white')

lgd = ax.set_legend(title='Vel. do Vento (m/s)', loc=(1.1, 0), shadow=True)

ax.set_yticks(np.arange(0, 25, step=5))

ax.set_yticklabels(np.arange(0, 25, step=5), fontsize='12')

plt.title("Rosa dos ventos",y=1.08, fontsize='20')

plt.setp(lgd.get_texts(), fontsize=18) # Aumenta o texto da legenda
```

```python
import pandas as pd

excel_file = 'input/SIMA_estreito.xlsx'
data = pd.read_excel(excel_file, index_col=[0], usecols="A,B,C")

data = data[::-1]
data
```

```python
data = data.rename(index=str, columns={"Direção do vento(°NV)": "dir",
                                       "Intensidade do vento(m/s)": "vel"});
data.index.names = ['date']

data.head()
```

```python
idx = pd.date_range('2006-02-12 17:00:00', '2007-01-29 14:00:00', freq='60T')
idx
```

```python
data.index = pd.DatetimeIndex(data.index)
from numpy import NaN

data = data.reindex(idx, fill_value=NaN)
data.tail()
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
plt.plot(kind='barh')
plt.plot(data.vel)
```

```python
plt.subplots()

plt.plot(data['vel'].groupby(data['vel'].index.month).mean())
```

```python
#plt.rcdefaults()
plt.xkcd()
plt.subplots()
plt.plot(data['vel'].groupby(data['vel'].index.hour).mean(),'-or', label='Estreito')

data_filtro = data[data['vel'] >= 5].dropna()

plt.plot(data_filtro['vel'].groupby(data_filtro['vel'].index.hour).mean(),'-o', label='Estreito Filtro')

plt.legend(bbox_to_anchor=(1.3, 1.02))
plt.title('Ciclo diurno médio da Velocidade do Vento')
plt.ylabel('Velocidade do Vento (m/s)')
plt.xlabel('Hora')
```
