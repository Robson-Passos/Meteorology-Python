```python
import pandas as pd

dados = pd.read_table('dados.txt', delim_whitespace=True, names=('A', 'B', 'vel', 'dir'))
```

```python
dados.head()
```

```python
from windrose import WindroseAxes, WindAxes, plot_windrose
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams.update({'font.size': 22})  # Isso  para aumentar o valor dos labels N, N-E, E, etc..., tambm aumenta o ttulo da legenda

ax = WindroseAxes.from_ax()

ax.bar(dados.dir, dados.B, normed=True, bins=np.arange(0,10,2), opening=0.8, edgecolor='white')

lgd = ax.set_legend(title='ndice de Turbulncia', loc=(1.1, 0), shadow=True)

ax.set_yticks(np.arange(0, 25, step=5))

ax.set_yticklabels(np.arange(0, 25, step=5), fontsize='12')

#plt.title("Rosa dos ventos",y=1.08, fontsize='20')

plt.setp(lgd.get_texts(), fontsize=18) # Aumenta o texto da legenda

plt.savefig.dpi: 1200 # Define o DPI da figura a ser salva

plt.savefig('Rosa dos ventos.png', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
